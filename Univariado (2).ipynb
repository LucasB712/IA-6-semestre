```python
import matplotlib.pyplot as plt  # ✅
import pandas as pd
import numpy as np

df = pd.read_csv("diabetes.csv")  # confira o nome do arquivo
```

```python
df.columns = [c.strip().replace(" ", "_") for c in df.columns]


idade = df["Age"].dropna()

# 1) Histograma
plt.figure(figsize=(6,4))
plt.hist(idade, bins=15, edgecolor="black", alpha=0.7)
plt.title("Histograma da Idade")
plt.xlabel("Idade")
plt.ylabel("Frequência")
plt.show()
```

```python
# 2) Boxplot
plt.figure(figsize=(4,6))
plt.boxplot(idade, vert=True, patch_artist=True)
plt.title("Boxplot da Idade")
plt.ylabel("Idade")
plt.show()
```

```python
# 3) Gráfico de Violino
plt.figure(figsize=(4,6))
plt.violinplot(idade, showmeans=True)
plt.title("Gráfico de Violino da Idade")
plt.ylabel("Idade")
plt.show()
```
